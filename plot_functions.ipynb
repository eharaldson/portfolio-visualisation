```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
def filter_orders_by_ticker(df_orders: pd.DataFrame, tickers: list[str]) -> pd.DataFrame:
    """
    Filters the orders DataFrame to include only the specified tickers.
    
    Parameters:
    df_orders (pd.DataFrame): DataFrame containing order data.
    tickers (list[str]): List of ticker symbols to filter by.
    
    Returns:
    pd.DataFrame: Filtered DataFrame containing only the specified tickers.
    """
    return df_orders[df_orders['ticker'].isin(tickers)]

def filter_orders_by_date(df_orders: pd.DataFrame, start_date: date, end_date: date) -> pd.DataFrame:
    """
    Filters the orders DataFrame to include only orders from a specified start date onwards.
    
    Parameters:
    df_orders (pd.DataFrame): DataFrame containing order data.
    start_date (date): The start date to filter orders from.
    
    Returns:
    pd.DataFrame: Filtered DataFrame containing orders from the specified start date onwards.
    """
    return df_orders[df_orders['date'] >= start_date & df_orders['date'] <= end_date]

def plot_tickers(df_orders: pd.DataFrame, tickers: list[str], start_date: date, end_date: date = None) -> pd.DataFrame:
    """
    """

    plt.figure(figsize=(14, 7))
    for ticker in tickers:
        ticker_data = df_orders[df_orders['ticker'] == ticker]
        sns.lineplot(data=ticker_data, x='date', y='price', label=ticker)

    plt.title('Stock Prices Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
